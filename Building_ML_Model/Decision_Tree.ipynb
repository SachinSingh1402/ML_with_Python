# Decision Tree

1. suprevised learning model
2. used for Classification and Regression
3. Builds Decision Nodes at each steps
4. Basis of Tree-based Models

Advantages:

1. can be used for both Classification & Regression
2. Easy to interpet
3. No need for normalization or scalling
4. Not sensitive to outlier


Disadvantages:

1. Overfitting issue
2. Small changes in the data alter the tree structure causing
3. Training time is relatively higher

<b>ENTROPY:</b>

Entropy is the quantitative measure of the <b>randomness</b> of the information being processed.

A <b>high value of Entropy</b> means that the <b>Randomnes</b> in the system is <b>high</b> and thus making accurate predicition is tough.

A <b>low value of Entropy</b> means that the <b>Randomnes</b> in the system is <b>low</b> and thus making accurate predicition is tough.



Information gain is the measure of how much information feature provides a class. Low entropy leads to increase information Gain and high entropy leads to low information gain.

information gain computes the difference between entropy before split and average entropy after split of the dataset based on a given feature.

GINI IMPURITY

The split made in a Decision Tree is said to be pure if all the data points are accurately separated into different classes.

Gini Impurity measure the likelihood that a randomly selected data points would be incorrectly classified by a specific node.
