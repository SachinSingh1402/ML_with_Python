# Pandas Data Frame :

Pandas DataFrame is two-dimensional tabular data structure with label axes(rows and colums)

```python
# importing the pandas library
import pandas as pd
import numpy as np
```

# Creating a Pandas Data Frame

```python
from sklearn.datasets import fetch_openml
boston = fetch_openml(name='boston', version=1)
```

```python
type(boston)
```

```python
print(boston)
```

# Pandas DataFrame

```python
boston_df = pd.DataFrame(boston.data, columns = boston.feature_names)
```

```python
boston_df.shape
```

```python
type(boston_df)
```

```python
# locating a particular column
print(boston_df.iloc[:,0])  # first column
print(boston_df.iloc[:,1])  # second column
print(boston_df.iloc[:,2])  # third column
print(boston_df.iloc[:,-1]) # last column
```
